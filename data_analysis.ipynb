Before we proceed to further steps, data cleaning is performed: replacing nan and non-float values in euro_values for all currencies.

## Data Cleaning

Check if the date column has nan values

```python
data["Metrics_Dates"][6:].isnull().values.any()
```

It is observed that the non-float values in euro_values are 'nan' and '-'. So we replace them by "1.00". A better and more realistic improvisation would be to choose the last date with a float value in the dataframe. For example, the euro_value of "AUD" on "2021-06-05" does not appear on the dataframe (actually, '-'). So, an algorithm could search the nearest date in the past with a (float) euro-value, say "2021-06-04" and assign (1,5804) the value to "2021-06-05."

```python
data = data.fillna("-")
```

```python
data = data.replace("-", "1.00")
```

```python
for col in data.columns[1:]:
    
    try:
        data[col][6:] = data[col][6:].apply(lambda x: float(str(x).replace(",", ".")))
    except:
        data[col][6:] = data[col][6:].apply(lambda x: str(x).replace(",", "."))

```

# Part 4: Create a table named fact_exchange_rate_history


This table is generated from the table (dataframe) dim_currency

```python
dates = dim_currency_df["last_updated_date"].unique()
currency_codes = dim_currency_df["curr_code"].unique()

  
```

```python
#function to generate euro value of currencies for a date

def generate_curr_euro_value_by_date(date):
    
    curr_from = dict.fromkeys(currency_codes, None)
    
    for key, value in curr_from.items():

        curr_from[key] = dim_currency_df[(dim_currency_df["last_updated_date"] == date) 
                                         & (dim_currency_df["curr_code"] == key)]["one_euro_value"].to_list()[0]

    return curr_from
```

```python
#function for exchange rate for one currency for a date
def exchange_rate_one_curr(base_curr, curr_euro_value_dict):
    result = {}
    for key, val in curr_euro_value_dict.items():
        if key != base_curr:
            result[key] = curr_euro_value_dict[base_curr]/val
    return result

```

## Upload the dataframe in batches by date (daily task)

```python
table_name = "fact_exchange_rate_history"
```

Let's see the table after one iteration over dates

```python
dummy = 0 
df_after_first_date_iter = pd.DataFrame()
```

```python
#loop over dates
for date in dates: 

    
    t_df = pd.DataFrame()
    
    #generate the euro value of currencies for each date
    currencies_euro_value = generate_curr_euro_value_by_date(date)
     
    
    # exchange rates between the currencies for a date
    for key, value in currencies_euro_value.items():
        
    
        tm_df = pd.DataFrame(
                {
                    "history_date": date,
                    "from_cur_code": key,
                    "to_cur_code": pd.Series(exchange_rate_one_curr(key, currencies_euro_value).keys()),
                    "exchange_rate": pd.Series(exchange_rate_one_curr(key, currencies_euro_value).values()),
                }
            )
        
        t_df = pd.concat([t_df, tm_df], ignore_index= True)
    
    ##======= not part of the algorithm: from here==========
    dummy +=1
    if dummy ==1:
        df_after_first_date_iter = t_df
    ##======= not part of the algorithm: to here==========
    
    #append the exchange table for a date to the BigQuery table    
    pandas_gbq.to_gbq(t_df, f"{dataset_id}.{table_name}", project_id=project_id, if_exists= 'append')
        
```
